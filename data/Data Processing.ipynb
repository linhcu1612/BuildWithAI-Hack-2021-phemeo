```python
import pandas as pd
from sklearn import preprocessing
```

```python
df = pd.read_csv("student info2.csv")
pd.set_option('display.max_columns', None)
```

```python
df.columns
```

```python
df = df.rename(columns={"How many hours per week do you spend on writing assignments?": "Hours_On_Assignment",\
                        "How many hours per week do you spend on studies?": "Hours_On_Studies",\
                        'How much time does it take for you to reach college?':'Travel_Time',\
                        'How much would you consider as your average attendance throughout all the semesters so far?':'Attendance',\
                        'How do you spend your free time?':'Free_Time',\
                        'Internet availability at home': 'Internet_Availability',\
                        'If yes, what is your internet speed?': 'Internet_Speed',\
                        'What is your preferred mode of transportation to reach college?':'Mode_Of_Transportation',\
                        'Do you prefer self - study or group study?':'Selfstudy_or_Groupstudy',\
                        'Do you do your submissions on time?':'On_Time_Submissions',\
                        'Have you enrolled for coaching classes?':'Coaching_classes',\
                        'How socially active are you?':'Social_Skills',\
                        'Can you sit a lecture for 2 hrs straight?': '2h_lecture',\
                        'If there are 5 hrs of lectures per day, would you attend all?': 'Attend_full_day_lectures',\
                        'If there are 5 hrs of practicals per day, would you attend all?': 'Attend_full_day_practices',\
                        'Have you won or participated in any Non - Technical event?': 'Non_Tech_Event', \
                        "Teacher's Feedback": 'Feedback'
                })
```

```python
df.columns
```

```python
df = df[['SEM 1 SGPA', 'SEM 2 SGPA', 'SEM 3 SGPA', 'SEM 4 SGPA', 'SEM 5 SGPA',
       'SEM 6 SGPA', 'SEM 7 SGPA', 'SEM 8 SGPA', 'Travel_Time', 'Attendance',
       'Internet_Speed', 'Mode_Of_Transportation', 'On_Time_Submissions',
       'Coaching_classes', 'Feedback', 'Label']]
df.columns
```

```python
df["Label"] = df["Label"].fillna(0)
df['Fast_Internet'] = df['Internet_Speed'] \
.replace({'High (>= 10 Mbps)': 1,
          'Low (< 10 Mbps)': 0}) \
.fillna(0)
df['Travel_Time'] = df['Travel_Time'] \
.replace({'Less than 1 hour': 'Short',
          '1 - 2 hours': 'Median', 
          '2 - 3 hours': 'Long', 
          'More than 3 hours': 'Long'})
df['Attendance'] = df['Attendance'] \
.replace([ 'Less than 60%', '60%  - 69%', '70%  - 79%', '80%  - 89%', '90%  - 100%'], \
        ['Poor', 'Normal', 'Normal', 'Good', 'Good'])
df['Attendance'] = df['Attendance'] \
.replace({'Less than 60%': 'Poor',
          '60%  - 69%': 'Normal',
          '70%  - 79%': 'Normal',
          '80%  - 89%': 'Good', 
          '90%  - 100%': 'Good'})
df['Public_Transport'] = df['Mode_Of_Transportation'] \
.replace({'Bus': 1, 
          'Train': 1, 
          'Private Vehicle': 0})
def check_feedback(feedback):
    lc_feedback = str(feedback).lower()
    positive = ["good", "better", "leadership", "responsible", "eagerness", "hard working", "excellent", "effort", "disciplined", "solid", "respectful", "self-motivated"]
    negative =["not", "poor", "disappointed", "late", "argue", "lagging", "bunk", "improvement", "talkative"]
    for neg in negative:
        if neg in lc_feedback:
            return 0
    for pos in positive:
        if pos in lc_feedback:
            return 1
    return feedback
df['Positive_Feedback'] = df['Feedback'].apply(check_feedback)
df['On_Time_Submissions'] = df['On_Time_Submissions'].replace({'Yes': 1, 'No': 0})
df['Coaching_classes'] = df['Coaching_classes'].replace({'Yes': 1, 'No': 0})
df = pd.get_dummies(df, columns=["Travel_Time", "Attendance"])
df = df.drop(['Feedback', 'Internet_Speed', 'Mode_Of_Transportation'], axis=1 )
for i in range(1, 9):
    col = f"SEM {i} SGPA"
    df[col] = df[col].fillna(df[col].median())
```
